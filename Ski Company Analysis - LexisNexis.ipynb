```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
columns = ['Number of Customers (N)','Current Ski','Faster Ski', '2nd Ski Lift']
data = pd.read_excel('ski.xlsx', header = None, names = columns)
```

```python
# Top 9 Summary
data = data.drop(labels=0, axis=0)
data = data.head(9)
data
```

```python
# Creating figure and axis objects using subplots()
fig, ax = plt.subplots(figsize=[16, 10])

# multiple line plots
# plotting the line 1 - 'Current Ski' points 
ax.plot(data['Number of Customers (N)'],data['Current Ski'], label = "Current Ski",marker='o', markerfacecolor='blue', 
        markersize=12, color='skyblue', linewidth=4, linestyle='dashed')

# plotting the line 2 - 'Faster Ski' points 
ax.plot(data['Number of Customers (N)'],data['Faster Ski'],label='Faster Ski',marker='o',markerfacecolor='olivedrab',
        markersize=12, color='yellowgreen', linewidth=4, linestyle='dashed')

# plotting the line 3 - '2nd Ski Lift' points 
ax.plot(data['Number of Customers (N)'],data['2nd Ski Lift'],label = '2nd Ski Lift',marker='o',
        markerfacecolor='rebeccapurple',markersize=12, color='mediumpurple', linewidth=4,linestyle='dashed')

# Set the x axis label of the current axis.
plt.xlabel('Number of Customers (N)',fontsize=15)

# Set the y axis label of the current axis.
plt.ylabel('Queue waiting time (mins)',fontsize=15)

# Set a title of the current axes.
plt.title('Comparing Ski Lift Methods',fontsize=30)

# show a legend on the plot
plt.legend()

# Display a figure.
plt.show()
```
